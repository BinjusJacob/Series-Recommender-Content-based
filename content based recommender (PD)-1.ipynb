```python
import re


# Define a function to clean each string
def clean_year_string(string):
    # Regular expression pattern to extract start and end years
    pattern = r'\b(\d{4})–?(\d{4})?\b'

    # Search for start and end years in the string
    match = re.search(pattern, string)

    if match:
        # Extract start and end years
        start_year = int(match.group(1))
        end_year = int(match.group(2)) if match.group(2) else 2024
    else:
        # Handle cases where no match is found
        start_year = 1955
        end_year = 2024

    return start_year, end_year

# Clean each string in the feature
cleaned_features = [clean_year_string(string) for string in imdb_df["Runtime_of_Series"]]

print(cleaned_features)
```

### Building recommender function

```python
indices=pd.Series(imdb_df.index,index=imdb_df["Series_Title"]).drop_duplicates()
```

```python
indices
```

what we will include in our recommender function:

    1. Declare the title of the movie as an argument
    2. Obtain the index of the movie from the indices reverse maping
    3. Get the list of cosine similarity scores for that particular movie with all movies using cosine_sim.Convert this into a list of tuples where the first element is the position and second element is the similarity score
    4. Sort this list of tuples on the basis of similarity scores
    5. Get the top 10 elements of the list .Ignore the first element as it is refers to the similarity score with itself.
    6.Return the titles corresponding to the indices of the top 10 elements.excluding the first

```python
#function  takes title as input 
def plt_content_recom(cosine_sim=cosine_sim,df=imdb_df,indices=indices):
        #Obtain the index of the movie that matches the input
        print("Input series title")
        Series_Title=input().capitalize()
        idx=indices[Series_Title]
        # get the pairwise similarity scores of all movies against  that movie and covert it into a list of tuples
        sim_scores=list(enumerate(cosine_sim[idx]))
        #sort movies based on the similarity
        sim_scores=sorted(sim_scores,key= lambda x: x[1],reverse=True)
        # most similar 10 movies,ignore the first movie
        sim_scores=sim_scores[1:11]
        #Get the movie indices
        movie_indices=[i[0] for i in sim_scores]
        #Return the top 10 most similar movies
        return imdb_df["Series_Title"].iloc[movie_indices]
```
